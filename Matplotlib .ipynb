<font size = '5'><b> MatplotLib Tutorial <font>

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

    .)Generally easy to get started for simple plots
    -Support for custom labels and texts
    -Great control of every element in a figure
    -High-quality output in many formats
    -Very customizable in general


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#Simple examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("just a 2D graph")
plt.savefig('Plot1')
```

```python
y = x**2
```

```python
#plot a staight line
plt.plot(x,y,'r+--',linewidth =2,markersize =1)
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title("2D diagra")
```

```python
#creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x**2,y,'r--')
```

```python
x = np.arange(1,10)
y = 0.4*x+10
plt.plot(x,y)
```

```python
#for plotting some sine waves
np.pi
    
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.title("Sine wave")
plt.show()
```

```python
#bar plots
x = [2,5,10]
y = [25,36,25]

plt.bar(x,y)
```

<font size = '5'><b> Histogram <font>

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

<font size = '5'><b> Box Plot using Matplotlib <font>

```python

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

<b>Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
